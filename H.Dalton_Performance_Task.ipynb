```python
#let's go back and create this dataframe again, but for ALL employee types

jobs = df4[['employee_number','american_indian_alaskan', 'asian', 'black_african_american', 'native_hawaiian_pacific', 'white','hispanic_latino_flag','gender','hours_per_week','position_name', 'hourly_salary' ]].copy()
jobs
```

```python
#Step 1- change ethnicity columns from string (Y/N) to integeer (1, 0)
jobs['asian'] = pd.Series(np.where(jobs.asian.values == 'Y', 1, 0),
          jobs.index)
jobs['american_indian_alaskan'] = pd.Series(np.where(jobs.american_indian_alaskan.values == 'Y', 1, 0),
          jobs.index)
jobs['black_african_american'] = pd.Series(np.where(jobs.black_african_american.values == 'Y', 1, 0),
          jobs.index)
jobs['native_hawaiian_pacific'] = pd.Series(np.where(jobs.native_hawaiian_pacific.values == 'Y', 1, 0),
          jobs.index)
jobs['white'] = pd.Series(np.where(jobs.white.values == 'Y', 1, 0),
          jobs.index)
jobs['hispanic_latino_flag'] = pd.Series(np.where(jobs.hispanic_latino_flag.values == 'Y', 1, 0),
          jobs.index)

jobs.dtypes
```

```python
#Step 2- Change ethnicity columns from integeer (1, 0) to boolean (true/false)

jobs.american_indian_alaskan = jobs.american_indian_alaskan.astype('bool')
jobs.asian = jobs.asian.astype('bool')
jobs.black_african_american = jobs.black_african_american.astype('bool')
jobs.native_hawaiian_pacific = jobs.native_hawaiian_pacific.astype('bool')
jobs.white = jobs.white.astype('bool')
jobs.hispanic_latino_flag = jobs.hispanic_latino_flag.astype('bool')


jobs.dtypes
```

```python
#Step 3- create new column called "non_minority?"
#true= employee identifies as white and only white
#false= employee does not identify as white, and identifies as ethnic/racial minority

jobs['non_minority?'] = (jobs['asian'] == False) & (jobs['american_indian_alaskan'] == False) & (jobs['black_african_american'] == False) & (jobs['native_hawaiian_pacific'] == False) & (jobs['hispanic_latino_flag'] == False)  
jobs
```

```python
#Step 4 create new df with the columns that are needed for this initial analysis
non_minority = jobs[['employee_number','gender','hours_per_week','position_name', 'hourly_salary', 'non_minority?' ]].copy()
non_minority
```
